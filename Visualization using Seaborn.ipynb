# Visualizing Data in Seaborn

## Python Libraries for Data Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
iris = pd.read_csv('https://d24c4belkzmc2t.cloudfront.net/cdn-uploads/1561115577_iris.csv') 
iris.head()
```

```python
wine_reviews = pd.read_csv('https://d24c4belkzmc2t.cloudfront.net/cdn-uploads/1561115763_winemag-data_first150k.csv',low_memory = False, nrows=1000)
wine_reviews.head()
```

## Creating Scatter Plot in Seaborn –

```python
sb.scatterplot(x = iris['sepal_length'], y = iris['sepal_width'])
```

## Creating Stripplot in Seaborn –

```python
df = sb.load_dataset('iris')
sb.stripplot(x = 'species', y = 'petal_length', data = df)
plt.show()
```

## Creating Swarmplot in Seaborn –

```python
df = sb.load_dataset('iris')
sb.swarmplot(x = 'species', y = 'petal_length', data = df)
plt.show()
```

## Creating a Line Plot in Seaborn –

```python
sb.lineplot(data = iris.drop(['class'], axis = 1))
```

## Creating Histogram in Seaborn –

```python
df = sb.load_dataset('iris')
sb.distplot(df['petal_length'], kde = False)
plt.show()
```

```python
df = sb.load_dataset('iris')
sb.distplot(df['petal_length'],bins = 10, kde = True)
plt.show()
```

## Creating Bar chart in Seaborn –

```python
sb.countplot(wine_reviews['points'])
```

## Creating Box plot in Seaborn –

```python
df = wine_reviews[(wine_reviews['points']>=95) & (wine_reviews['points'] <= 100)]
sb.boxplot('points','price',data = df)
plt.ylim(0,1000)
```

## Creating Violin plot in Seaborn –

```python
df = wine_reviews[(wine_reviews['points']>=95) & (wine_reviews['points'] <= 100)]
sb.violinplot('points','price',data = df)
plt.ylim(0,800)
```

## Creating Heatmap in Seaborn –

```python
sb.heatmap(iris.corr(), annot = True)
```

## Creating Pair plot in Seaborn –

```python
sb.pairplot(iris)
```

```python
from pandas.plotting import scatter_matrix

fig,ax = plt.subplots(figsize = (12,12))
scatter_matrix(iris, alpha = 1, ax = ax)
plt.show()
```
